```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train
```

```python
from sklearn import preprocessing
```

```python
scaler=preprocessing.StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
```

```python
a=LinearRegression()
b=SVR()
c=RandomForestRegressor()
d=GradientBoostingRegressor()
e=AdaBoostRegressor()
reg_list=[a,b,c,d,e]

```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

score={}

for reg in reg_list:
    reg.fit(x_train,y_train)
    predict=reg.predict(x_test)
    print('-'*50)
    print(reg)
    print(f"R2_score:{r2_score(y_test, predict)}")
    print(f"Mean Absolute Error Score : {mean_absolute_error(y_test,predict)}")
    print(f"Mean Sqaured Error Score : {mean_squared_error(y_test,predict)}")
    score[reg]={
        "R2 Score":r2_score(y_test,predict),
        "Mean Absolute Error Score":mean_absolute_error(y_test,predict),
        "Mean Squared Error Score":mean_squared_error(y_test,predict)
    }
          
```
